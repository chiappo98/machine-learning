# Explore your data with visualization

We will make:

* ***Univariate*** plots
   * Histograms
   * Density Plots
   * Box and Whisker Plots
* ***Multivariate*** plots

## Import the data

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/dbonacorsi/AMLBas2122/main/datasets/pima-indians-diabetes.data.csv'

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(url, names=names)
data
```

## Univariate plots

### Histograms

```python
from matplotlib import pyplot
```

```python
data.hist()
pyplot.rcParams["figure.figsize"] = [16,16]
pyplot.show()
```

### Density plots

```python
from matplotlib import pyplot
```

```python
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
pyplot.show()
```

### Box and Whisker plots

```python
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
pyplot.show()
```

## Multivariate plots
They show the interactions between multiple variables in your dataset.

### Correlation matrix plots

```python
from matplotlib import pyplot
import numpy
```

```python
correlations = data.corr()
correlations
```

```python
# plot correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```

### Scatter plot matrix

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(data)
pyplot.show()
```
